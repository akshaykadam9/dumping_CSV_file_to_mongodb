```python
import pandas as pd
from pymongo.mongo_client import MongoClient
class mb:
    def __init__(self,file_path,con,database_name,document_name):
        try:
            self.path=file_path
            self.con=con
            self.database_name=database_name
            self.document_name=document_name
            self.connection=MongoClient(self.con)
            self.db=self.connection[self.database_name]
            self.document=self.db[self.document_name]
        except Exception as e:
            print(e)
    
        
    def read_file(self):
        """ using this function we can read csv file. """
        try:
            df = pd.read_csv(self.path, sep=';')
            return df
        except Exception as e:
            print(e)
    def dict_record(self):
        '''using this function we are able to convert csv data ro dictonary format '''
        try:
            b = self.read_file().to_dict(orient='records')
            return b
        except Exception as e:
            print(e)
            
    def dump_data(self):
        '''By this function we are able to dump a data into mongodb databse documents'''
        try:
            self.document.insert_many(self.dict_record())
            print('successfully dump')
        except Exception as e:
            print(e)
    def delete_data(self):
        '''By this function we are able to delete all the record in Mongodb databse documents'''
        try:
            self.document.delete_many({})
            print('data is deleted')
        except Exception as e:
            print(e)
            
    def find_tuples(self):
        '''By this funtion we are able to see the records of Mongodb databse documents'''
        try:
            c=self.document.find()
            return c
        except Exception as e:
            print(e)
    def find_specific_tuples(self,query):
        '''By this function we are able to find specific records in mongodb databse documents for th that we have to provide a query'''
        try:
            c=self.document.find(query)
            return c
        except Exception as e:
            print(e)
    def update_tuple(self,query,set_query):
        '''By this function we are able to update a specific records that we want to update for that provide query and subquery'''
        try:
            c=self.document.update_many(query,set_query)
            print('records updated succesfully')
        except Exception as e:
            print(e)
```
